# Gathering Data

The data I chose for my project was the stroke prediction dataset. I chose it because it has a high usability and mostly numerical values, this suits a decision tree classifier model.

```python
train_file_path = '../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv'

# Create a new Pandas DataFrame with our training data
stroke_train_data = pd.read_csv(train_file_path)

#stroke_test_data.columns
stroke_train_data.describe(include='all')
#stroke_train_data.head()
```

# Preparing the Data

Our prediction data is whether or not a patient had a stroke, so we must tell the program to predict this and separate it from the rest. We separate 'y' (our prediction target). We will also make sure that there aren't any missing values.


# Choosing Data

Now, I will select which data I will want to consider when predicting strokes. I will only be removing the 'ID' data column because it is not relevant to our prediction. I will also remove some data which is categorical in an attempt to minimise the dataset to only the most relevant features that aren't categorical. The target data is included, for now.

```python
# Let's reduce our data to only the features we need and the target.
# The features we chose have similar 'count' values when we describe() them
# We need to keep the target as part of our DataFrame for now.
selected_columns = ['gender', 'age', 'hypertension', 'heart_disease', 
                    'avg_glucose_level', 'bmi', 'smoking_status', 'stroke']

# Create our new training set containing only the features we want
prepared_data = stroke_train_data[selected_columns]

# Drop rows (axis=0) that contain missing values
prepared_data = prepared_data.dropna(axis=0)

# Check that you still have a good 'count' value. The value should be the same for all columns.
# If your count is very low then you may need to remove features with the lowest count.
prepared_data.describe(include='all')
```

# Separating Features from Target

We will separate the features used for prediction from the target data so that our model can learn how to predict strokes.

```python
# Separate out the prediction target
y = prepared_data.stroke

# Drop the target column (axis=1) from the original dataframe and use the rest as our feature data
X = prepared_data.drop('stroke', axis=1)

# Take a look at the data again
X.head()
#y.head()
```

# One Hot Encoding

Some features, such as gender and smoking status are not numerical in our dataset. The easiest way to fix this issue is to encode them as numbers. I will use the get_dummies function to do so.

```python
# One hot encode the features. This will only act on columns containing non-numerical values.
one_hot_X = pd.get_dummies(X)

one_hot_X.head()
```

There are now more columns in our data, for the different categories of these values. 

# Separating Training and Testing Data

It is important we separate our prediction dataset into two sections; training (which will take up the majority of the data) and testing which will validate what the model has learned. We cannot test the model on the same data we trained it on as it may memorise the answers, picking up patterns specific to the testing data.

```python
# Dataset is separated into 'train' and 'val' sections
train_X, val_X, train_y, val_y = train_test_split(one_hot_X, y, random_state = 1)
```

# Choosing and Training a Model

The Decision Tree Classifier is the best fit for this task, as it can take many inputs and will output a categorical prediction. Our categorical predictions will be either 0, for someone who has not had a stroke and 1, for someone who has had a stroke.

Here we train and graph the results.

```python
stroke_predictor = DecisionTreeClassifier(max_depth=4)

# Train the model on the one hot encoded data
stroke_predictor.fit(train_X, train_y)

# Let's plot the tree to see what it looks like!
plt.figure(figsize = (20,10))
plot_tree(stroke_predictor,
          feature_names=train_X.columns,
          class_names=['no stroke', 'stroke'],
          filled=True)
plt.show()
```

# Evaluating Model and Tuning Hyperparameters

Now that our model is nearing completion, we must test its performance and see if we can make it any better by making some small changes.

```python
print("Making predictions for the first 5 patients in the training set.")

# Get the first five predictions as a list
pred = stroke_predictor.predict(one_hot_X)

print("The predictions are:")

# Merge actual target values and predictions back in with original features to see how we went.
X['Stroke'] = y
X['Predicted'] = pred

X.head()
```

Not looking good, but that isn't a very accurate representation of how good our model is, as it is part of the training data.

Let's also test the mean absolute error of our model.

```python
predictions = stroke_predictor.predict(val_X)

print(mean_absolute_error(val_y, predictions))
```

# Investigating Further...

Not bad, only making wrong predictions about 5% of the time.

But in the data we looked at, it was always wrong. A bit odd...

What if we make the decision tree a bit bigger?

Let's make the max depth 16 and see how it performs.

```python
stroke_predictor = DecisionTreeClassifier(max_depth=16)

# Train the model on the one hot encoded data
stroke_predictor.fit(train_X, train_y)

# Let's plot the tree to see what it looks like!
plt.figure(figsize = (20,10))
plot_tree(stroke_predictor,
          feature_names=train_X.columns,
          class_names=['no stroke', 'stroke'],
          filled=True)
plt.show()
```

```python
predictions = stroke_predictor.predict(val_X)

print(mean_absolute_error(val_y, predictions))
```

Now its doing worse...

A likely reason for this is because about 5% of the data is of people who have had a stroke, so simply predicting that nobody in the dataset has had a stroke will produce decent results. 

But in real-life, this isn't useful as high-risk patients will not be informed. A better approach is to have a model that predicts whether someone is high-risk by prioritising saving lives rather than prioritising being accurate. If we predict someone will have a stroke but they don't have one, not much can go wrong. If we predict someone will not have a stroke when they will, everything can go wrong.

```python
print("Making predictions for the first 5 patients in the data set.")

# Get the first five predictions as a list
pred = stroke_predictor.predict(one_hot_X)

print("The predictions are:")

# Merge actual target values and predictions back in with original features to see how we went.
X['Stroke'] = y
X['Predicted'] = pred

X.head()
```

This is what happens to the mean absolute error when we increase the max depth (the y-axis should say "Mean Absolute Error" but I mistyped).

What if we test how well it predicts out of the people who will have a stroke? We can use the start of the dataset as a good estimate.

As it turns out, sklearn does have a feature which can only test positive predictions, called precision. Lets test it. 

```python
stroke_predictor = DecisionTreeClassifier(max_depth=21)

# Train the model on the one hot encoded data
stroke_predictor.fit(train_X, train_y)

# Let's plot the tree to see what it looks like!
plt.figure(figsize = (20,10))
plot_tree(stroke_predictor,
          feature_names=train_X.columns,
          class_names=['no stroke', 'stroke'],
          filled=True)
plt.show()

predictions = stroke_predictor.predict(val_X)

print(precision_score(val_y, predictions))
```

Here is how it changes over time when we alter the max tree depth:

At lower depths, the tree will just predict that there were no strokes every single time, meaning there was a prediction score of zero. After that, the prediction score converges at about 15%... not exactly what I was hoping for.

```python
# Let's reduce our data to only the features we need and the target.
# The features we chose have similar 'count' values when we describe() them
# We need to keep the target as part of our DataFrame for now.
selected_columns = ['Age', 'Sex', 'RestingBP', 'FastingBS', 'RestingECG',
                    'ExerciseAngina', 'ST_Slope', 'HeartDisease']

# Create our new training set containing only the features we want
prepared_data = heart_failure_train_data[selected_columns]

# Drop rows (axis=0) that contain missing values
prepared_data = prepared_data.dropna(axis=0)

# Check that you still have a good 'count' value. The value should be the same for all columns.
# If your count is very low then you may need to remove features with the lowest count.
prepared_data.describe(include='all')
```
